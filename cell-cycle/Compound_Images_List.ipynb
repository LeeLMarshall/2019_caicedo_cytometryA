```python
plates = numpy.unique(df.Image_PathName_OrigHoechst.values)
# plates = [plates[0]]
plates
```

## Compound selection and creation of the list of the corresponding images

```python
compounds = ['Gabazine', 'CYCLOPIAZONIC ACID', 
             'ESTRADIOL', 'DMSO', 
             'Proglumide', 'PIZOTIFEN', 
             'CATECHIN PENTABENZOATE', 'TRIADIMEFON',
             '3-HYDROXYCOUMARIN', 'Etoposide',
             'DACTINOMYCIN', 'Colchicine',
             'Blebbistatin']


compound = compounds[0]
if compound == 'DMSO':
    codf1 = df[df.Image_Metadata_ASSAY_WELL_ROLE == 'mock']
    
    for ind, plate in enumerate(plates) :
        codf2 = codf1[codf1.Image_PathName_OrigHoechst == plate]
        l = len(codf2)
        rows = range(50)
        codf2 = codf2.iloc[rows]
        
        if ind == 0:
            indicies = list(codf2.index)
            continue
            
        indicies += list(codf2.index)
        codf = codf1.loc[indicies]
    
else:
    codf = df[df.Image_Metadata_SOURCE_COMPOUND_NAME.str.lower() == compound.lower()]

codf.head(10)
```

```python
coimages = codf[['Image_FileName_OrigHoechst',
               'Image_PathName_OrigHoechst']]

coimages
```

```python
image_list = pd.DataFrame(columns = ['DNA', 'Done'])

image_list
```

```python
for i in range(0, len(coimages.index)):
    row = coimages.index[i]
#     print('"{}"'.format(coimages.iloc[i].Image_FileName_OrigHoechst))
    ipath = 'BBBC022_v1_images_' 
    ipath += coimages.iloc[i].Image_PathName_OrigHoechst[:-1] 
    ipath += 'w' 
    ipath += coimages.iloc[i].Image_FileName_OrigHoechst[16]
    ipath += '/'
    ipath += coimages.iloc[i].Image_FileName_OrigHoechst
#     print("Path of index {} : {}".format(row, ipath))
    image_list.loc[i] = [ipath, 'False']
    
image_list
```
